```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os
import re
```

```python
path = "pubmedHSE.txt"
```

```python
df = pd.read_table(path)
```

```python
df
```

```python
IDs = []
for item in df['1']:
    if 'PMID' in item:
        IDs.append(item.split(" ")[1])
    
```

```python
len(IDs)
```

```python
citations = []
for number in IDs:
    try:
        path = os.path.join("https://pubmed.ncbi.nlm.nih.gov/", number)
        page = requests.get(path)
        soup = BeautifulSoup(page.content, 'html.parser')
        cited = soup.find_all('h2')[7].get_text().split("        ")[2]
        citations.append(re.escape(cited).split("\\")[0])
    except:
        print(path)
citations


# backup = citations2
```

```python
for i in range(len(citations)):
    citations[i] = int(citations[i])
sum(citations)
```
